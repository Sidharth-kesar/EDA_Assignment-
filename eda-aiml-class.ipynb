```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
#
```

```python
df = pd.read_csv('/kaggle/input/new-dataset/zomato.csv',encoding = 'latin-1')
df.head()
```

```python
sns.heatmap(df.isnull(),yticklabels=False,cbar=True,cmap='viridis')
```

```python
df_country = pd.read_excel('/kaggle/input/new-dataset/Country-Code.xlsx')
df_country.head()
```

```python
final_df = pd.merge(df,df_country , on='Country Code', how='left')
final_df.head(2)
```

```python
final_df.dtypes
```

```python
final_df.columns
```

```python
country_names = final_df.Country.value_counts().index
country_val = final_df.Country.value_counts().values
```

```python
plt.pie(country_val[:3],labels=country_names[:3],autopct='%1.9f%%')
```

```python
ratings=final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'rating count'})
```

```python
ratings
```

```python
plt.rcParams['figure.figsize']=(12,6)
sns.barplot(x='Aggregate rating', y='rating count', data=ratings)
```

```python
sns.countplot(x='Rating color',data = ratings,palette=['blue','yellow','green','orange','red'])
```
